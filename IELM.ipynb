We then code the regression training and testing functions.

```
def trainRegressionUCI(train_input_data,outputLabel):  
    outputLabel = sig(outputLabel)
    lmax = 20
    E = outputLabel
    l = 0
    defError = 0.1 
    for l in range(1,lmax):
        hMat = np.zeros(shape = (train_input_data.shape[0],l))
        for row in train_input_data:
           k=0
           row = np.reshape(row, (1,np.product(row.shape)))
           h = []
           for i in range(l):
               weights = np.random.random((train_input_data.shape[1],1))
               output = predVal(row,weights) 
               h.append(output)   
           h = np.reshape(h, (-1, l))  
           hMat[k] = h[0]
           k = k + 1
        beta = np.dot(np.linalg.pinv(hMat), outputLabel)
        E = E - np.dot(hMat,beta)
        E_VAL = math.sqrt( np.square(np.subtract(outputLabel,E)).mean() )
        if(E_VAL < defError) :
            print("Training RMSE: ",E_VAL)
            break 
    return beta

def testRegressionUCI(data, outputD,b):
    outputD =sig(outputD)
    la = b.shape[0]
    hMat = np.zeros(shape = (data.shape[0],la))
    for row in data:
       k=0
       row = np.reshape(row, (1,np.product(row.shape)))
       h = []
       for i in range(la):
           weights = np.random.random((data.shape[1],1))
           output = predVal(row,weights)
           h.append(output)
       h = np.reshape(h, (-1, la))  
       hMat[k] = h[0]
       k = k + 1 
    o = sig(np.dot(hMat , b))
    rmse =  math.sqrt( np.square(np.subtract(outputD,o)).mean() )
    print("Testing RMSE",rmse)
```

```
import numpy as np
import pandas as pd
import math
import time as t
b = 0.01
def predVal(inputs,w):
    inputs = inputs.astype(float)
    output = relu(np.dot(inputs, w) + b) 
    return output

def sig(x):
    calculate = 1/(1+np.exp(-x))
    return calculate

def relu(x):
    cal = np.maximum(x, 0)
    return cal

def predAcc(originalLabel,predicted):
        matched = 0
        for i in range(len(originalLabel)):
                if originalLabel[i] == predicted[i]:
                    matched += 1
        accuracyVal = matched / float(len(originalLabel))      
        return accuracyVal

def trainClassificationMNIST(train_input_data,outputLabel):  
    
    lmax = 20
    E = outputLabel
    l = 0
    Acc = 0.999
    for l in range(1,lmax):
        hMat = np.zeros(shape = (train_input_data.shape[0],l))
        for row in train_input_data:
           k=0
           row = np.reshape(row, (1,np.product(row.shape)))
           h = []
           for i in range(l):
               weights = np.random.random((train_input_data.shape[1],1))
               output = predVal(row,weights)
               h.append(output)
           h = np.reshape(h, (-1, l))  
           hMat[k] = h[0]
           k = k + 1
        beta = np.dot(np.linalg.pinv(hMat), outputLabel)
        E = E - np.dot(hMat,beta)
        E = relu(E)
        E_VALC = predAcc(outputLabel,E)
        if(E_VALC > Acc) :
            print("Training Accuracy: ",E_VALC*100,"%")
            break
    return beta

def testClassificationMNIST(data, outputD,b):
    la = b.shape[0]
    hMat = np.zeros(shape = (9999,la))
    for row in data:
       k=0
       row = np.reshape(row, (1,np.product(row.shape)))
       h = []
       for i in range(la):
           weights = np.random.random((784,1))
           output = predVal(row,weights)
           h.append(output)
       h = np.reshape(h, (-1, la)) 
       hMat[k] = h[0]
       k = k + 1  
    o = relu(np.dot(hMat , b))
    acc = predAcc(outputD, o)
    print("Testing Accuracy",acc*100,"%")    

#-----------------------------------------------Classification-----------------------------------------------
start=t.time()
b123=trainClassificationMNIST(d1[0:19999,1:785],d1[0:19999,0:1])
end=t.time()
print("(Classification) MNIST training time:",end-start," seconds\n")
start=t.time()
testClassificationMNIST(d2[0:9999,1:785],d2[0:9999,0:1],b123)
end=t.time()
print("(Classification) MNIST testing time:",end-start," seconds\n")
#------------------------------------------------Regression---------------------------------------------------
start=t.time()
beta123=trainRegressionUCI(traffic_train[0:24000,0:4],traffic_train[0:24000,4:5])
end=t.time()
print("(Regression) UCI training time:",end-start," seconds\n")
start=t.time()
testRegressionUCI(traffic_test[0:24203,0:4],traffic_test[0:24203,4:5],beta123)
end=t.time()
print("(Regression) UCI testing time:",end-start," seconds\n")
```
